```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
import time
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
```

```python
driver = webdriver.Chrome(executable_path = '/Users/DM/FS/final/chromedriver')
```

```python
def get_person_details():
    person = {}
    try:
        d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[2]/div[1]/h3/span/span[1]')
        person['headline'] = d.text
    except:
        person['headline'] = None
#     try:
#         driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[2]/div[2]/o-profile-overview[2]/div/p/span/a').click()
#         d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[2]/div[2]/o-profile-overview[2]/div/p')
#         person['bio'] = d.text
#     except:
#         try:
#             d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[2]/div[2]/o-profile-overview[2]/div/p/span/span[1]')
#             person['bio'] = d.text
#         except:
#             person['bio'] = None
    try:
        d = driver.find_element_by_xpath('//*[@id="oProfilePage"]/div[1]/div[2]/cfe-profile-skills-integration/div/div/section')
        person['skills'] = [item.text for item in d.find_elements_by_tag_name('span')]
    except:
        person['skills'] = None
    try:
        d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[3]/ul/li[1]/div[1]/div/h3/cfe-profile-rate/span/span')
        person['rate'] = d.text
    except:
        person['rate'] = None
    try:
        d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[3]/ul/li[2]/h3/span')
        person['earned'] = d.text
    except:
        person['earned'] = None
    try:
        d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[3]/ul/li[3]/h3')
        person['jobs'] = d.text
    except:
        person['jobs'] = None
    try:
        d = driver.find_element_by_xpath('//*[@id="optimizely-header-container-default"]/div[3]/ul/li[4]/h3')
        person['hours'] = d.text
    except:
        person['hours'] = None
    return person
```

```python
def page_of_ten_scrape(first = True):
    people = []
    if first == True:
        begin = 2
        end = 12
    else:
        begin = 1
        end = 11
    for i in range(begin, end):
        try:
            driver.find_element_by_xpath('//*[@id="layout"]/div[2]/div[2]/div/div/div/div/div/freelancers-search/div/div[2]/section[' + str(i) + ']/div/div/article/div/div[1]/div/div[2]/div[1]/a/span').click()
            time.sleep(1)
            p = get_person_details()
            print("trying to get person " + str(i))
            people.append(p)
            driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/button').click()
            time.sleep(1)
        except:
            pass
    return people
```

```python
def get_all():
    peeps = []
#     pp = page_of_ten_scrape(True)
#     peeps += pp
#     time.sleep(2)
    i = 0
    while i < 500:
        i += 1
        print("trying new page")
        try:
            driver.find_element_by_xpath('//*[@id="layout"]/div[2]/div[2]/div/div/div/div/div/freelancers-search/div/footer/div/ul/li[10]/a/span[1]').click()
            time.sleep(2)
            pp = page_of_ten_scrape(False)
            peeps += pp
            time.sleep(2)
        except:
            break
    return peeps
```
